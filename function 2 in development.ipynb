```python
def parse_ebird(file_name):
    import csv
    f = open(file_name,'r')
    data = csv.reader(f,delimiter='\t')
    lat = []
    long = []
    date = []
    for row in data:
        lat.append(row[22])
        long.append(row[23])
        date.append(row[24])
    #there is some crap at the beginning and end of these lists to remove
    lat = lat[1:-2]
    long = long[1:-2]
    date = date[1:-2]
    return lat, long, date
```

```python
lat, long, date = parse_ebird("trbl_ebird.txt")

# problem:  the following code turns these data into a really nice dictionary
# with years as keys and lat, long as values.  However, the dates are not
# unique and data gets lost.  How to resolve...
```
