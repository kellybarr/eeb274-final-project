```python
def parse_ebird(file_name):
    import csv
    f = open(file_name,'r')
    data = csv.reader(f,delimiter='\t')
    lat = []
    long = []
    date = []
    for row in data:
        lat.append(row[22])
        long.append(row[23])
        date.append(row[24])
    #there is some crap at the beginning and end of these lists to remove
    lat = lat[1:-2]
    long = long[1:-2]
    date = date[1:-2]
    return lat, long, date
```

```python
lat, long, date = parse_ebird("trbl_ebird.txt")

```

```python
def making_dict_ebird(lat, long, date):
    #making dates unique so data is not lost
    date2 = []
    for i in range(len(date)):
        date2.append(date[i] + "_" + str(i))
    bird_data = {}
    bird_data = dict(zip(date2,zip(lat,long)))
    return bird_data
```
